```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
#Generate random observations from gaussian distributions
std_normal=np.random.normal(loc=0,scale=1,size=1000)
#loc -> mean, scale--> variance, size--> 100 such samples
```

```python
for i in range(1,101):
    print(i,np.percentile(std_normal,i))
```

```python
#generate 100 samples from N(20,5)
measurements=np.random.normal(loc=20,scale=5,size=100)
#try size 1000, as no of samples increases more and more points will lie on the line
#If no of samples is small Q-Q plot is hard to interpret (Limitation)

stats.probplot(measurements,dist="norm",plot=pylab) # probploy-->probability plot # dist=norm // N(0,1)
pylab.show()
#X axis--> N(0,1) and #Y-->measurements
```

```python
#generate 100 samples from N(20,5)
measurements=np.random.uniform(low=-1,high=1,size=1000)
#try size 1000,

stats.probplot(measurements,dist="norm",plot=pylab) 
pylab.show()

# On x axis I have a gaussian ditribution, on y axis I have a uniform distribution
```
